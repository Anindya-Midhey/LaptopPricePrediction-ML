```python
# --- Step 1: Clean the 'Memory' column ---
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)
df['Memory'] = df['Memory'].str.replace('GB', '', regex=False)
df['Memory'] = df['Memory'].str.replace('TB', '000', regex=False)

# --- Step 2: Split into two layers ---
new = df['Memory'].str.split('+', n=1, expand=True)
df['first'] = new[0].str.strip()
df['second'] = new[1].fillna('0')  # fill missing second parts

# --- Step 3: Create Layer indicators BEFORE removing letters ---
df['Layer1HDD'] = df['first'].apply(lambda x: 1 if 'HDD' in x else 0)
df['Layer1SSD'] = df['first'].apply(lambda x: 1 if 'SSD' in x else 0)
df['Layer1Hybrid'] = df['first'].apply(lambda x: 1 if 'Hybrid' in x else 0)
df['Layer1Flash_Storage'] = df['first'].apply(lambda x: 1 if 'Flash Storage' in x else 0)

df['Layer2HDD'] = df['second'].apply(lambda x: 1 if 'HDD' in x else 0)
df['Layer2SSD'] = df['second'].apply(lambda x: 1 if 'SSD' in x else 0)
df['Layer2Hybrid'] = df['second'].apply(lambda x: 1 if 'Hybrid' in x else 0)
df['Layer2Flash_Storage'] = df['second'].apply(lambda x: 1 if 'Flash Storage' in x else 0)

# --- Step 4: Now strip out non-numeric characters safely ---
df['first'] = df['first'].astype(str).str.replace(r'\D', '', regex=True).replace('', '0').astype(int)
df['second'] = df['second'].astype(str).str.replace(r'\D', '', regex=True).replace('', '0').astype(int)

# --- Step 5: Compute total storage by type ---
df['HDD'] = (df['first'] * df['Layer1HDD']) + (df['second'] * df['Layer2HDD'])
df['SSD'] = (df['first'] * df['Layer1SSD']) + (df['second'] * df['Layer2SSD'])
df['Hybrid'] = (df['first'] * df['Layer1Hybrid']) + (df['second'] * df['Layer2Hybrid'])
df['Flash_Storage'] = (df['first'] * df['Layer1Flash_Storage']) + (df['second'] * df['Layer2Flash_Storage'])

# --- Step 6: Drop temporary helper columns ---
df.drop(columns=['first', 'second','Layer1HDD', 'Layer1SSD', 'Layer1Hybrid', 'Layer1Flash_Storage','Layer2HDD', 'Layer2SSD', 'Layer2Hybrid', 'Layer2Flash_Storage'], inplace=True)
```

```python
import time
import random
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso,ElasticNet
from sklearn.metrics import mean_squared_error, r2_score , mean_absolute_error
from sklearn.preprocessing import LabelEncoder, StandardScaler , PolynomialFeatures
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
label_encoders = {}
obj_cols = X.select_dtypes(include='object').columns
for col in obj_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    label_encoders[col] = le

print("Label encoding applied to categorical columns.")
```

```python
# Split the transformed data into training and test sets

# First split: 70% train, 30% temp (validation + test)
X_train, X_temp, Y_train, Y_temp = train_test_split(X, Y, test_size=0.3, random_state=42)

# Second split: 15% validation, 15% test (out of total)
X_val, X_test, Y_val, Y_test = train_test_split(X_temp, Y_temp, test_size=0.50, random_state=42)
```

```python
# Calculate performance metrics
def calculate_performance_metrics(Y_test, Y_pred):
    # Calculate R-squared
    r2 = r2_score(Y_test, Y_pred)
    # Calculate Mean Absolute Error
    mae = mean_absolute_error(Y_test, Y_pred)
    # Calculate Mean Squared Error
    mse = mean_squared_error(Y_test, Y_pred)
    return r2, mae, mse
```

```python
def k_fold_evaluation(model, X, Y, cv=5):
    kf = KFold(n_splits=cv, shuffle=True, random_state=42)
    mse_scores = -cross_val_score(model, X, Y, cv=kf, scoring='neg_mean_squared_error')
    mae_scores = -cross_val_score(model, X, Y, cv=kf, scoring='neg_mean_absolute_error')
    r2_scores = cross_val_score(model, X, Y, cv=kf, scoring='r2')
    print(f"\nK-Fold Validation ({cv}-fold):")
    print(f"Average MSE: {mse_scores.mean():.6f}")
    print(f"Average MAE: {mae_scores.mean():.6f}")
    print(f"Average R2 Score : {r2_scores.mean():.6f}")
    return mse_scores.mean(), mae_scores.mean(), r2_scores.mean()
```

```python
# Create and train the model
t0 = time.time()
model_lr = LinearRegression()
model_lr.fit(X_train, Y_train)
# --- Validation evaluation ---
Y_val_pred = model_lr.predict(X_val)
r2_val_lr, mae_val_lr, mse_val_lr = calculate_performance_metrics(Y_val, Y_val_pred)
print("\nValidation Results (Linear Regression)")
print(f"MSE : {mse_val_lr:.6f}")
print(f"MAE : {mae_val_lr:.6f}")
print(f"R2 score  : {r2_val_lr:.6f}")

# --- Test evaluation ---
Y_pred_lr = model_lr.predict(X_test)
r2_lr, mae_lr, mse_lr = calculate_performance_metrics(Y_test, Y_pred_lr)
print("\nTest Results (Linear Regression)")
print(f"MSE : {mse_lr:.6f}")
print(f"MAE : {mae_lr:.6f}")
print(f"R2 score  : {r2_lr:.6f}")

k_fold_evaluation(model_lr, X, Y)
t1 = time.time()
runtime_lr = t1- t0
print( f"\nRuntime: {runtime_lr:.6f} seconds")
```

```python
# Initialize and train the Ridge Regression model
t0 = time.time()
ridge_reg = Ridge(alpha=0.7)
ridge_reg.fit(X_train, Y_train)
# --- Validation ---
Y_val_pred_ridge = ridge_reg.predict(X_val)
r2_val_ridge, mae_val_ridge, mse_val_ridge = calculate_performance_metrics(Y_val, Y_val_pred_ridge)
print("\nValidation Results (Ridge Regression)")
print(f"MSE : {mse_val_ridge:.6f}")
print(f"MAE : {mae_val_ridge:.6f}")
print(f"R2 score  : {r2_val_ridge:.6f}")

# --- Test ---
Y_pred_ridge_reg = ridge_reg.predict(X_test)
r2_ridge_reg, mae_ridge_reg, mse_ridge_reg = calculate_performance_metrics(Y_test, Y_pred_ridge_reg)
print("\nTest Results (Ridge Regression)")
print(f"MSE : {mse_ridge_reg:.6f}")
print(f"MAE : {mae_ridge_reg:.6f}")
print(f"R2 score  : {r2_ridge_reg:.6f}")
k_fold_evaluation(ridge_reg, X, Y)
t1 = time.time()
runtime_ridge = t1- t0
print( f"\nRuntime: {runtime_ridge:.6f} seconds")
```

```python
# Initialize and train the Lasso Regression model
t0 = time.time()
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(X_train, Y_train)

# --- Validation ---
Y_val_pred_lasso = lasso_reg.predict(X_val)
r2_val_lasso, mae_val_lasso, mse_val_lasso = calculate_performance_metrics(Y_val, Y_val_pred_lasso)
print("\nValidation Results (Lasso Regression)")
print(f"MSE : {mse_val_lasso:.6f}")
print(f"MAE : {mae_val_lasso:.6f}")
print(f"R2 score  : {r2_val_lasso:.6f}")

# --- Test ---
Y_pred_lasso_reg = lasso_reg.predict(X_test)
r2_lasso_reg, mae_lasso_reg, mse_lasso_reg = calculate_performance_metrics(Y_test, Y_pred_lasso_reg)
print("\nTest Results (Lasso Regression)")
print(f"MSE : {mse_lasso_reg:.6f}")
print(f"MAE : {mae_lasso_reg:.6f}")
print(f"R2 score  : {r2_lasso_reg:.6f}")
k_fold_evaluation(lasso_reg, X, Y)
t1 = time.time()
runtime_lasso = t1- t0
print( f"\nRuntime: {runtime_lasso:.6f} seconds")
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import KFold

# Model names
models_kfold = [
    'Linear Regression',
    'Ridge Regression',
    'Lasso Regression',
    'Elastic Net Regression',
    'Polynomial Regression (Degree 2)',
    'Decision Tree Regressor',
    'Random Forest Regressor'
]

mse_kfold, mae_kfold, r2_kfold = [], [], []

# ---- Linear Regression ----
mse_lr_k, mae_lr_k, r2_lr_k = k_fold_evaluation(LinearRegression(), X, Y, cv=5)
mse_kfold.append(mse_lr_k); mae_kfold.append(mae_lr_k); r2_kfold.append(r2_lr_k)

# ---- Ridge ----
mse_ridge_k, mae_ridge_k, r2_ridge_k = k_fold_evaluation(Ridge(alpha=0.7), X, Y, cv=5)
mse_kfold.append(mse_ridge_k); mae_kfold.append(mae_ridge_k); r2_kfold.append(r2_ridge_k)

# ---- Lasso ----
mse_lasso_k, mae_lasso_k, r2_lasso_k = k_fold_evaluation(Lasso(alpha=0.001), X, Y, cv=5)
mse_kfold.append(mse_lasso_k); mae_kfold.append(mae_lasso_k); r2_kfold.append(r2_lasso_k)

# ---- Elastic Net ----
mse_elastic_k, mae_elastic_k, r2_elastic_k = k_fold_evaluation(
    ElasticNet(alpha=0.01, l1_ratio=0.001), X, Y, cv=5)
mse_kfold.append(mse_elastic_k); mae_kfold.append(mae_elastic_k); r2_kfold.append(r2_elastic_k)

# ---- Polynomial Regression (deg 2) ----
model_poly = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
mse, mae, r2 = k_fold_evaluation(model_poly, X, Y, cv=5) # X is not scaled here, this might cause issues for poly deg 3 if the scaler expects already scaled data.
mse_kfold.append(mse); mae_kfold.append(mae); r2_kfold.append(r2)

# ---- Descision Tree Regressor ----
mse_dt_k, mae_dt_k, r2_dt_k = k_fold_evaluation(
    DecisionTreeRegressor(random_state=42), X, Y, cv=5)
mse_kfold.append(mse_dt_k); mae_kfold.append(mae_dt_k); r2_kfold.append(r2_dt_k)

# ---- Random Forest Regressor ----
mse_rf_k, mae_rf_k, r2_rf_k = k_fold_evaluation(RandomForestRegressor(
    n_estimators=100, random_state=42), X, Y, cv=5)
mse_kfold.append(mse_rf_k); mae_kfold.append(mae_rf_k);
r2_kfold.append(r2_rf_k)



# ================== Create DataFrame ==================
kfold_results = pd.DataFrame({
    'Model': models_kfold,
    'K-Fold MSE': mse_kfold,
    'K-Fold MAE': mae_kfold,
    'K-Fold R2 score': r2_kfold
}).round(6)

# ================== Highlight Best Models ==================
def highlight_best(s):
    """Highlight min for errors and max for R2 score."""
    is_r2 = 'R²' in s.name or 'R2' in s.name
    if is_r2:
        best = s.max()
        return ['background-color: lightgreen; font-weight: bold' if v == best else '' for v in s]
    else:
        best = s.min()
        return ['background-color: lightgreen; font-weight: bold' if v == best else '' for v in s]

# Ensure the column name matches the DataFrame
styled_table = kfold_results.style.apply(highlight_best, subset=['K-Fold MSE', 'K-Fold MAE', 'K-Fold R2 score'])

# ================== Display Table ==================
print("\n K-Fold Cross-Validation Results (Average of 5 Folds):\n")
display(styled_table)
```
